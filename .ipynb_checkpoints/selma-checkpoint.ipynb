```python

def find_id(x):
    if type(x) is str and "immoweb" in x:
        for n in re.findall("([0-9]+)",x):
            if len(n) == 7:
               
                return int (n)
    else: 
        return x
print (find_id(df.iloc[(27931,1)]))
print ((df.iloc[(27931,1)]))
print (df.iloc[30424, 2])

print (find_id(df.iloc[(77771,3)]))
```
