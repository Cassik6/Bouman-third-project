```python
import pandas as pd
import numpy as np
import matplotlib

# visualization libraries
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns



# print the graphs in the notebook
% matplotlib inline

# set seaborn style to white
sns.set_style("white")
# set panda options for float big price
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df = pd.read_csv('immo_data_set.csv')
df
```

```python
print()
appart = df[df['house_is'] == 0]
print(appart.price.max())
appart_outlier = sns.boxplot(x=appart['price'])
appart_outlier.set(xlim=(10000, 5600000))

```

```python
house = df[df['house_is'] == 1]
print(house.price.max())
print(house.price.min())
house_outlier = sns.boxplot(x=house['price'])
house_outlier.set(xlim=(10000, 24000000.0))
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(appart['area'], appart['price'])
ax.set_xlabel('Area in m²')
ax.set_ylabel('Price of the appartment')
fmt = '{x:,.2f}€'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 
plt.show()


```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(house['area'], house['price'])
ax.set_xlabel('Area in m²')
ax.set_ylabel('Price of the house')
fmt = '{x:,.2f}€'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 
plt.show()
```
